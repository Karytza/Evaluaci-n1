```python
#Crea columnas del mes, día, hora, minuto, segundo

df3["AÑO"]=pd.DatetimeIndex(df3["FECHA"]).year
df3["MES"]=pd.DatetimeIndex(df3["FECHA"]).month
df3["DÍA"]=pd.DatetimeIndex(df3["FECHA"]).day
df3["HORA"]=pd.DatetimeIndex(df3["FECHA"]).hour
df3["MINUTO"]=pd.DatetimeIndex(df3["FECHA"]).minute
df3["SEGUNDO"]=pd.DatetimeIndex(df3["FECHA"]).second
df3["TEMPERATURA"]=pd.Index(df3["AirTC_Avg"])
```

```python
#Crear una lista como función de otras columnas
df=pd.DataFrame({})
df["TEMPERATURA"]=df3.groupby(["AÑO","MES","DÍA","HORA"])["TEMPERATURA"].transform("mean")
df.head()
```

```python
#Crear nuevas columnas en base a las temperaturas máximas y mínimas
df["TMIN"]=df3.groupby(["AÑO","MES","DÍA"])["TEMPERATURA"].transform("min")
df["TMAX"]=df3.groupby(["AÑO","MES","DÍA"])["TEMPERATURA"].transform("max")
df["HORA"]=df3.HORA
df["DIA"]=df3.DÍA
df["MES"]=df3.MES
df["AÑO"]=df3.AÑO
```

```python
df.head()
```

```python
df = df.drop_duplicates(subset=['AÑO','MES','DIA','HORA'])
df=df.reset_index(drop=True)
df.head()
```

```python
#Calculamos el índice UF24 (acumulación de horas frío desde el primer día)
count=[]
for i in range (0,len(df)):
    z=df.TEMPERATURA[i]
    if (z<=1.4) :
        j=0
    if (z>1.4 and z<=2.4) :
        j=0.5   
    if (z>2.4 and z<=9.1) :
        j=1
    if (z>9.1 and z<=12.4) :
        j=0.5  
    if (z>12.4 and z<=15.9) :
        j=0
    if (z>15.9 and z<=18) :
        j=-0.5
    if (z>18) :
        j=-1
    count.append(j)
df["UF24"]=count

df.head()
```

```python
#Graficar la TMAX y TMIN
df1=df.drop(["TEMPERATURA", "HORA", "UF24"],1)
df1=df1.drop_duplicates()
f=len(df1)
df1.index = np.arange(0,f)
df1.head()


sns.set(rc={'figure.figsize':(12,8.27)})

df4 = df1[['TMAX', "TMIN"]]
fig=plt.figure(); df4.plot(); plt.legend(loc='best')
plt.title("Evolución de las temperaturas mínima y máxima")
plt.ylabel("Temperatura ºC")
plt.grid(True)
plt.plot(label = "TMAX", color = '#D2691E')   
plt.plot(label = "TMIN", color = '#1E90FF')  


plt.show()
```

```python
"Crear gráfica de la evolución de las temperaturas máximas y mínimas: suma acumulada UF24"
df3=df.drop(["TEMPERATURA", "TMIN","TMAX"],1)
df3=df3.drop_duplicates()
f=len(df3)
df3.index = np.arange(0,f)
df3.head()

df3["SUF"]=df3["UF24"].cumsum()

df4 = df3[["SUF"]]
plt.figure(figsize=(20,19))
fig=plt.figure(); df4.plot(); plt.legend(loc='best')
plt.title("Evolución de la suma acumulada UF24")
plt.ylabel("Temp ºC")
plt.grid(True)

plt.show()
```
